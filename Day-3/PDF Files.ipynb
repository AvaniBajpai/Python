#### Extracting Metadata
We can extract some of the important useful data from any pdf. For example, we can extract the information on author of the document, its title, subject and the number of pages contained in the pdf file.

```python
from PyPDF2 import PdfFileReader
```

```python
#import os
```

```python
def extract_pdfMeta(path):
    with open(path,'rb') as f:
        pdf = PdfFileReader(f) #extract text data from the interactive PDF in Python
        info = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()
    print("Author: \t",info.author)
   
    print("Creator: \t",info.creator)
    
    print("Producer: \t",info.producer)
   
    print("Subject: \t",info.subject)
    
    print("title: \t",info.title)
   
    print("Number of Pages in pdf: \t",number_of_pages)
    #return info
path = 'D://Files//DSA.pdf'
extract_pdfMeta(path)
```

#### Extracting Text from PDFs

```python
def extract_pdfText(path):
    with open(path, 'rb') as f:
        pdf = PdfFileReader(f)
      # get the 51st page
        page = pdf.getPage(50)
        print(page)
        print('Page type: {}'.format(str(type(page))))
      #Extract text from the 50th page
        text = page.extractText()
        print(text)

path = 'C://Users//Nithya//Downloads//DSA.pdf'
extract_pdfText(path)
```

```python
### Rotate
from PyPDF2 import PdfFileWriter
#PythonFile = open('C://Users//Nithya//Downloads//Python.pdf', 'rb')
PythonFile = open('C://Users//Nithya//Downloads//DSA.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(PythonFile)
page = pdfReader.getPage(0)
page.rotateClockwise(90)
```

```python
pdfWriter = PyPDF2.PdfFileWriter()
pdfWriter.addPage(page) # add pages in a PDF document in Python
resultPdfFile = open('D://Files//rotatedPage.pdf', 'wb')
pdfWriter.write(resultPdfFile)
```
